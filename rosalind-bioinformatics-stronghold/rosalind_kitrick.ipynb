#### 1. Counting DNA nucleotides

```python
def nuc_count(seq_filepath):
    # open sequence file
    file = open(seq_filepath, 'r')
    # save singular sequence as string
    s = (file.readline()).strip('\n')
    # close file after reading
    file.close()
    # count instances of each nucleotide
    a = s.count('A')
    c = s.count('C')
    g = s.count('G')
    t = s.count('T')
    # format counts with spaces in between
    counts = str(a) + ' ' + str(c) + ' ' + str(g) + ' ' + str(t)
    # return each count in order
    return(counts)
```

```python
print(nuc_count('rosalind_dna.txt'))
```

 #### 2. Transcribing DNA into RNA

```python
def transcription(seq_filepath):
    # open sequence file
    file = open(seq_filepath, 'r')
    # save singular sequence as string
    t = (file.readline()).strip('\n')
    # close file
    file.close()
    # replace all instances of T in t with U
    rna = t.replace('T', 'U')
    # return RNA string
    return(rna)
```

```python
print(transcription('rosalind_rna.txt'))
```

#### 3. Complementing a strand of DNA

```python
def rev_complement(seq_filepath):
    # open sequence file
    file = open(seq_filepath, 'r')
    # save singular sequence as string
    s = (file.readline()).strip('\n')
    # close file
    file.close()
    # initializing dictionary for complementary nucleotides
    comp_nucs = {'A':'T','T':'A','C':'G','G':'C'}
    # reverse string of DNA
    reverse = s[::-1]
    # taking the complement of each nucleotide
    complement = ''
    for base in reverse:
        complement += comp_nucs[base]
    # return reverse complement
    return(complement)
```

```python
print(rev_complement('rosalind_revc.txt'))
```

#### 4. Rabbits and recurrence relations

```python
# this function will process the data from the file
def rabbit_recurrence(rabbit_filepath):
    # open file with n and k
    file = open(rabbit_filepath, 'r')
    # save data from file in list
    data = (file.readline()).split(' ')
    # close file
    file.close()
    # n : number of months
    n = int(data[0])
    # k : added rabbit pairs
    k = int(data[1])
    # call rabbit ribonacci function to solve
    return(rabbit_fibonacci(n,k))
# this function is able to be called recursively using n and k
def rabbit_fibonacci(n, k):
    # base cases for fibonacci 
    if n == 0:
        return(0)
    if n == 1:
       return(1)
    # recurssively solve using modified fibbonaci
    else:
       return(rabbit_fibonacci(n-1, k) + k*rabbit_fibonacci(n-2, k))
```

```python
print(rabbit_recurrence('rosalind_fib.txt'))
```

```python
def gc_content(fasta_filepath):
    # initialize empty dictionary to store IDs and GC content
    seq_content = {}
    # initialize empty sequence
    sequence = ''
    # open fasta file and iterate through lines
    file = open(fasta_filepath, 'r')
    for line in file:
        # if line is fasta header,
        if line.startswith('>'):
            # calculate gc content for previous sequence
            if sequence != '':
                gc_content = ((sequence.count('G') + sequence.count('C')) / len(sequence)) * 100
                seq_content[ID] = round(gc_content, 6)
            # save next sequence ID
            ID = line[1:14]
            # reset sequence to be empty for next
            sequence = ''
        # else, calculate gc content of sequence
        else:
            sequence += line.strip('\n')
    # calculate gc content for last sequence
    if sequence: 
        gc_content = ((sequence.count('G') + sequence.count('C')) / len(sequence)) * 100
        seq_content[ID] = round(gc_content, 6)
    # close file 
    file.close()
    # find ID with maximum value
    max_ID = max(seq_content, key=seq_content.get)
    max_GC = seq_content[max_ID]
    # return ID and max GC content
    return(max_ID + '\n' + str(max_GC))
```

#### 6. Counting Point Mutations

```python
def hamming_distance(seq_filepath):
    # open file with sequences
    file = open(seq_filepath, 'r')
    # iterate through file and save sequences to list
    seqs = []
    for line in file:
        seqs.append(line.strip('\n'))
    # close file
    file.close()
    # initialize hamming distance to be 0
    d = 0
    # iterate through index / length of strings
    for i in range(len(seqs[0])):
        # if nucleotides in strings are not equal
        if seqs[0][i] != seqs[1][i]:
            # increase distance by 1
            d += 1
    # return hamming distance
    return d
```

#### 7. Translating RNA into protein

```python
codon_table = {
    'AUG': 'M', 'UUU': 'F', 'UUC': 'F', 'UUA': 'L', 'UUG': 'L', 'UCU': 'S',
    'UCC': 'S', 'UCA': 'S', 'UCG': 'S', 'UAU': 'Y', 'UAC': 'Y', 'UGU': 'C',
    'UGC': 'C', 'UGG': 'W', 'CUU': 'L', 'CUC': 'L', 'CUA': 'L', 'CUG': 'L',
    'CCU': 'P', 'CCC': 'P', 'CCA': 'P', 'CCG': 'P', 'CAU': 'H', 'CAC': 'H',
    'CAA': 'Q', 'CAG': 'Q', 'CGU': 'R', 'CGC': 'R', 'CGA': 'R', 'CGG': 'R',
    'AUU': 'I', 'AUC': 'I', 'AUA': 'I', 'ACU': 'T', 'ACC': 'T', 'ACA': 'T',
    'ACG': 'T', 'AAU': 'N', 'AAC': 'N', 'AAA': 'K', 'AAG': 'K', 'AGU': 'S',
    'AGC': 'S', 'AGA': 'R', 'AGG': 'R', 'GUU': 'V', 'GUC': 'V', 'GUA': 'V',
    'GUG': 'V', 'GCU': 'A', 'GCC': 'A', 'GCA': 'A', 'GCG': 'A', 'GAU': 'D',
    'GAC': 'D', 'GAA': 'E', 'GAG': 'E', 'GGU': 'G', 'GGC': 'G', 'GGA': 'G',
    'GGG': 'G', 'UAA': 'Stop', 'UAG': 'Stop', 'UGA': 'Stop'
}
def translation(rna_file):
    # load sequence from txt file into string
    file = open(rna_file, 'r')
    # initialize empty string s for rna sequence
    rna = ''
    for line in file:
        rna += line.strip('\n')
    # close file
    file.close()
    # intialize protein sequence as empty string
    prot_seq = ''
    # iterate through nucleotides in RNA string with a step of 3
    for i in range(0, len(rna), 3):
        # store codon as three nucleotides
        codon = rna[i] + rna[i+1] + rna[i+2]
        # if codon is a stop codon, return prot_seq
        if codon_table[codon] == 'Stop':
            return prot_seq
        # else, append amino acid to prot_seq
        prot_seq += codon_table[codon]
    return prot_seq
```

```python
print(translation('rosalind_prot.txt'))
```

#### 8. Finding a motif in DNA

```python
def find_motif(seq_filepath):
    # open file for reading
    file = open(seq_filepath, 'r')
    # iterate through file and add sequences to seq list
    seqs = []
    for line in file:
        seqs.append(line.strip('\n'))
    # close file
    file.close()
    # s is the sequence being searched
    s = seqs[0]
    # t is the substring being searched for
    t = seqs[1]
    # initialize list of substring indexes
    locations = []
    # iterate through s, but stop once end of string is longer than t
    for i in range(len(s) - len(t) + 1):
        # if substring of len(t) starting at i matches t,
        if s[i:(i+len(t))] == t:
            # add starting index+1 (for nonzero indexing) to locations
            locations.append(i+1)
    # format locations with spaces in between
    locs = ''
    for place in locations:
        locs += str(place) + ' '
    # return locations of t as substring of s
    return(locs)
```

```python
print(find_motif('rosalind_subs.txt'))
```

#### 9. Finding a shared motif

```python
def longest_common_substring(fasta_filepath):
    # initialize empty list to store sequences
    seqs = []
    # initialize empty string to store sequence
    sequence = ''
    # open fasta file and iterate through lines
    file = open(fasta_filepath, 'r')
    for line in file:
        # if line is not a fasta header, 
        if line.startswith('>'):
            # if sequence is not empty
            if sequence != '':
                seqs.append(sequence)
            # reset sequence to be empty for next
            sequence = ''
        # else, calculate gc content of sequence
        else:
            sequence += line.strip('\n')
    # close file
    file.close()
    # sort seqs based on length to use shortest string first
    seqs.sort(key=len)
    # use shortest string to generate potential substring
    shortest = seqs[0]
    longest_common = ""
    # generate longest possible substring and then work downwards
    for length in range(len(shortest), 0, -1):
        for start in range(len(shortest) - length + 1):
            substring = shortest[start:start + length]
            # if substring is in other sequences in seqs
            if all(substring in sequence for sequence in seqs[1:]):
                # return that substring as LCS
                return substring 
    # if no substring found, return empty string
    return longest_common
```

#### 10.RNA Splicing

```python
codon_table = {
    'AUG': 'M', 'UUU': 'F', 'UUC': 'F', 'UUA': 'L', 'UUG': 'L', 'UCU': 'S',
    'UCC': 'S', 'UCA': 'S', 'UCG': 'S', 'UAU': 'Y', 'UAC': 'Y', 'UGU': 'C',
    'UGC': 'C', 'UGG': 'W', 'CUU': 'L', 'CUC': 'L', 'CUA': 'L', 'CUG': 'L',
    'CCU': 'P', 'CCC': 'P', 'CCA': 'P', 'CCG': 'P', 'CAU': 'H', 'CAC': 'H',
    'CAA': 'Q', 'CAG': 'Q', 'CGU': 'R', 'CGC': 'R', 'CGA': 'R', 'CGG': 'R',
    'AUU': 'I', 'AUC': 'I', 'AUA': 'I', 'ACU': 'T', 'ACC': 'T', 'ACA': 'T',
    'ACG': 'T', 'AAU': 'N', 'AAC': 'N', 'AAA': 'K', 'AAG': 'K', 'AGU': 'S',
    'AGC': 'S', 'AGA': 'R', 'AGG': 'R', 'GUU': 'V', 'GUC': 'V', 'GUA': 'V',
    'GUG': 'V', 'GCU': 'A', 'GCC': 'A', 'GCA': 'A', 'GCG': 'A', 'GAU': 'D',
    'GAC': 'D', 'GAA': 'E', 'GAG': 'E', 'GGU': 'G', 'GGC': 'G', 'GGA': 'G',
    'GGG': 'G', 'UAA': 'Stop', 'UAG': 'Stop', 'UGA': 'Stop'
}

def rna_splicing(fasta_filepath) -> str:
    # initialize empty list to store sequences
    seqs = []
    # initialize empty string to store sequence
    sequence = ''
    # open fasta file and iterate through lines
    file = open(fasta_filepath, 'r')
    for line in file:
        # if line is not a fasta header, 
        if line.startswith('>'):
            # if sequence is not empty
            if sequence != '':
                seqs.append(sequence)
            # reset sequence to be empty for next
            sequence = ''
        # else, calculate gc content of sequence
        else:
            sequence += line.strip('\n')
    # add final sequence after exiting the loop
    if sequence != '':
        seqs.append(sequence)
    # close file
    file.close()
    # isolate dna sequence from seqs list
    dna = seqs[0]
    # form list of introns from remaining seqs
    introns = seqs[1:]
    # for each intron in list,
    for substring in introns:
        # cut out indexes [start : start+len(substring)] from s
        new_dna = dna.replace(substring, '')
        # assign new_dna to be dna to continue if other introns
        dna = new_dna
    # after splicing out introns, convert DNA to RNA by replacing T with U
    rna = dna.replace('T','U')
    # translate rna to protein using codon_table
    # same method as translation(), but different input (file vs rna)
    # intialize protein sequence as empty string
    prot_seq = ''
    # iterate through nucleotides in RNA string with a step of 3
    for i in range(0, len(rna), 3):
        # store codon as three nucleotides
        codon = rna[i] + rna[i+1] + rna[i+2]
        # if codon is a stop codon, return prot_seq
        if codon_table[codon] == 'Stop':
            return prot_seq
        # else, append amino acid to prot_seq
        prot_seq += codon_table[codon]
    # return protein sequence
    return(prot_seq)
```

```python
print(rna_splicing('rosalind_splc.txt'))
```
